```python
#Using sqlite3 library database has been create
import sqlite3

#Using 'with' statement,database will be close automatically after every completed query 
with sqlite3.connect('Sql.db') as db:
    cursor = db.cursor()

#MYPY Phone Sytem table has been created 
cursor.execute('''CREATE TABLE IF NOT EXISTS MYPY (fullname TEXT , phonenum int)''')
#Database consists following coloumns; phone number and full name 


#Until loop interrupt by user,it will continue to work
while True:
    
    print("#"*30+"\nMYPY PHONE BOOK\n"+"#"*30)

    print("1 : Add New Entry\n"+"2 : Delete Entry\n"+"3 : Update Entry\n"+"4 : Lookup Number\n"+"5 : QUIT\n")

    operation = input("Which operation would you like to do: ")
    
    #Using try/except condition,you can be sure user enter valid number
    try:

        #Add Operation
        if operation == "1":
            
            fname = input("Please enter person fullname: ")
            number = int(input("Please enter telephone number:"))
            #Before add a new contact,check if it already exists.
            cursor.execute("SELECT * FROM MYPY WHERE phonenum = '{}'".format(number))
            db.commit()
            data = cursor.fetchall()  #Using fetchall(), we can assign results into 'data'
            
            #if data has an element,it means that the number is already assigned to user   
            if len(data) != 0:
                print(data)
                print("Number already exists.")
                
            else:
                cursor.execute("""INSERT INTO MYPY VALUES('{}','{}') """.format(fname, number))
                print("{} has been added into MYPY library with following number {}".format(fname, number))
                db.commit()
        
        #Delete Operation
        elif operation == "2":

            number2 = int(input("Please enter the number that you are going to delete: "))
            #Check whether number exists in MYPY sytem or not 
            cursor.execute("SELECT * FROM MYPY WHERE phonenum = '{}'".format(number2))
            db.commit()
            data = cursor.fetchall()
            print(data)
            
            if len(data) == 0:
                print("User has not been found.")
            
            else:
                
                cursor.execute("DELETE FROM MYPY WHERE phonenum='{}'".format(number2))
                db.commit()
                print("The user has been deleted")

        #Update Operation
        elif operation == "3":

            fname2 = input("Please enter name that you are going to update: ")
            number1 = int(input("Please enter the number that you are going to update: "))
            cursor.execute("UPDATE MYPY SET phonenum='{}'  WHERE fullname = '{}'".format(number1, fname2))
            db.commit()
            print("{}'s number has been updated with following number {}".format(fname2,number1))
       
        #Looking Up Operation
        elif operation == "4":
            
            number4 = int(input("Please enter the number that you are going to looking up: "))
            cursor.execute("SELECT * FROM MYPY WHERE phonenum = '{}'".format(number4))
            db.commit()
            data1 = cursor.fetchall()
            print(data1)
            
            #If 'data1' is  empty, programme provides user add a new record option
            if len(data1) == 0:
                
                print("The number could not be found.Would you like to insert a record?")
                
                a=input("Y/N:")
                
                if a == "Y":
                    
                    fname3 = input("Please enter person fullname: ")
                    cursor.execute("""INSERT INTO MYPY VALUES('{}','{}') """.format(fname3, number4))
                    print("{} has been added into MYPY library with following number {}".format(fname3, number4))
                    db.commit()
                    
                else:
                    print("Goodbye")
                    break
            
            #If 'data1' has a record then print record details 
            else:
                cursor.execute("SELECT fullname FROM MYPY WHERE phonenum = '{}'".format(number4))
                db.commit()
                h=cursor.fetchall()
                person = [x[0] for x in h]
                print("This number belongs to {}.".format(person[0]))

        #Quit Operation         
        elif operation == "5":
            break

        #If the operation out of range
        else:
            print("Invalid choice please check the operation value again")
            
    #If the operation out of range
    except ValueError:
        print("\nPlease enter valid number:")
        
```
